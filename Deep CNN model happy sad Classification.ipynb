```python
data= data.map(lambda x, y: (x/255,y))
```

```python
data.as_numpy_iterator().next()
```

```python
len(data)
```

```python
train_size= int(len(data)*0.7)
val_size= int(len(data)*0.2)
test_size= int(len(data)*0.1)
```

```python
train_size
```

```python
train= data.take(train_size)
val= data.skip(train_size).take(val_size)
test= data.skip(train_size+val_size).take(test_size)
```

```python
len(train)
```

```python
from tensorflow import keras
```

```python
model= keras.Sequential([
    keras.layers.Conv2D(16, (3,3), 1, activation='relu', input_shape= (256,256,3)),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Conv2D(32, (3,3), 1, activation='relu', input_shape= (256,256,3)),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Conv2D(32, (3,3), 1, activation='relu', input_shape= (256,256,3)),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
             loss=tf.losses.BinaryCrossentropy(),
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir= 'Perflogs'
tensorflow_callback= tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist= model.fit(train, epochs=20, validation_data=val, callbacks=[tensorflow_callback])
```

```python
fig= plt.figure(figsize=(10,10))
fig.suptitle("Loss", fontsize=20)
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
plt.legend(loc= 'upper left')
plt.show()
```

```python
fig= plt.figure(figsize=(10,10))
fig.suptitle("Accuracy", fontsize=20)
plt.plot(hist.history['accuracy'], color='teal', label='Accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='Val_Accuracy')
plt.legend(loc= 'upper left')
plt.show()
```

```python
model.evaluate(test)
```

```python
from tensorflow.keras.metrics import Precision, Recall, Accuracy, BinaryAccuracy
```

```python
pre= Precision()
rec= Recall()
acc= BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    x, y= batch
    yhat = model.predict(x)
    pre.update_state(y, yhat)
    rec.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f"Precision: {pre.result().numpy()},Recall:{rec.result().numpy()}, Accuracy: {acc.result().numpy()}")
```
